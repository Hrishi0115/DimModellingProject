```python
import pandas as pd 
```

```python
list_of_dict = []

with open(r'..\VillageCraftsSocialSalesDataV1.txt') as file:
    for row in file:
        lst = row[:-1].split(',')
        dct = {}
        for pair in lst:
            key, value = pair.split(':')
            key = key.lstrip().rstrip()
            value = value.lstrip().rstrip()
            dct[key] = value 
        list_of_dict.append(dct)


list_of_dict[0]
```

```python
df = pd.DataFrame(list_of_dict)

display(df)
```

```python
# clean dataframe 

df['Sale Price'] = df['Sale Price'].apply(lambda x: x[1:])
```

```python
display(df)
```

```python
df.to_csv(r'..\cleanedVillageCraftsSocialSalesData.csv', index=False)
```

```python
df.to_excel(r'..\cleanedVillageCraftsSocialSalesData.xlsx', index=False)
```
