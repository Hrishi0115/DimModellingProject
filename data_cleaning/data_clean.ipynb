```python
import pandas as pd 

df = pd.read_csv(r'..\VillageCraftsSocialSalesDataV1.txt')


display(df)
```

```python
list_of_dict = []

with open(r'..\VillageCraftsSocialSalesDataV1.txt') as file:
    for row in file:
        lst = row[:-1].split(',')
        dct = {}
        for pair in lst:
            key, value = pair.split(':')
            dct[key] = value 
        list_of_dict.append(dct)


list_of_dict[0]
```

```python
df = pd.DataFrame(list_of_dict)

display(df)
```
